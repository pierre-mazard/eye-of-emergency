# Comparaison des performances des modèles
Ce notebook compare les scores des différents modèles (DecisionTree, LogisticRegression, RandomForest, SVM, xgboost) sur les jeux d'entraînement, validation et test.

```python
# Import des librairies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Définir les modèles et les chemins des fichiers CSV
modeles = {
    'DecisionTree': {
        'train': '../results/decisiontree/metrics/decisiontree_train_predictions_summary.csv',
        'val': '../results/decisiontree/metrics/decisiontree_val_predictions_summary.csv',
        'test': '../results/decisiontree/metrics/decisiontree_test_predictions_summary.csv'
    },
    'LogisticRegression': {
        'train': '../results/logistic_regression/metrics/logistic_train_predictions_summary.csv',
        'val': '../results/logistic_regression/metrics/logistic_val_predictions_summary.csv',
        'test': '../results/logistic_regression/metrics/logistic_test_predictions_summary.csv'
    },
    'RandomForest': {
        'train': '../results/RandomForest/metrics/randomforest_train_predictions_summary.csv',
        'val': '../results/RandomForest/metrics/randomforest_val_predictions_summary.csv',
        'test': '../results/RandomForest/metrics/randomforest_test_predictions_summary.csv'
    },
    'SVM': {
        'train': '../results/SVM/metrics/svm_train_predictions_summary.csv',
        'val': '../results/SVM/metrics/svm_val_predictions_summary.csv',
        'test': '../results/SVM/metrics/svm_test_predictions_summary.csv'
    },
    'XGBoost': {
        'train': '../results/xgboost/metrics/xgboost_train_predictions_summary.csv',
        'val': '../results/xgboost/metrics/xgboost_val_predictions_summary.csv',
        'test': '../results/xgboost/metrics/xgboost_test_predictions_summary.csv'
    }
}
```

```python
# Charger tous les résultats dans un DataFrame global
dfs = []
for modele, splits in modeles.items():
    for split, path in splits.items():
        if os.path.exists(path):
            df = pd.read_csv(path)
            # On suppose que le CSV contient une seule ligne de scores
            scores = df.iloc[0][['accuracy', 'precision', 'recall', 'f1_score', 'roc_auc']].to_dict()
            scores['modele'] = modele
            scores['split'] = split
            dfs.append(scores)
df_all = pd.DataFrame(dfs)
```

```python
# Sélectionner les métriques principales pour la comparaison
metrics = ['accuracy', 'precision', 'recall', 'f1_score', 'roc_auc']
df_melted = df_all.melt(id_vars=['modele', 'split'], value_vars=metrics, var_name='metric', value_name='score')
```

```python
# Visualisation comparative
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x='modele', y='score', hue='split', ci=None)
plt.title('Comparaison des scores par modèle et split')
plt.ylabel('Score')
plt.xlabel('Modèle')
plt.legend(title='Split')
plt.show()
```

```python
# Visualisation détaillée par métrique
for metric in ['accuracy', 'precision', 'recall', 'f1_score', 'roc_auc']:
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_melted[df_melted['metric'] == metric], x='modele', y='score', hue='split', ci=None)
    plt.title(f'Comparaison {metric} par modèle et split')
    plt.ylabel(metric)
    plt.xlabel('Modèle')
    plt.legend(title='Split')
    plt.show()
```

```python
# Générer une image finale récapitulative de tous les modèles et splits
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
metrics = ['accuracy', 'precision', 'recall', 'f1_score', 'roc_auc']
for i, metric in enumerate(metrics):
    ax = axes[i//3, i%3]
    sns.barplot(data=df_melted[df_melted['metric'] == metric], x='modele', y='score', hue='split', ci=None, ax=ax)
    ax.set_title(f'{metric} par modèle et split')
    ax.set_ylabel(metric)
    ax.set_xlabel('Modèle')
    ax.legend(title='Split')
fig.suptitle('Synthèse des performances des modèles')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('../results/comparaison/figures/comparaison_modeles_synthese.png')
plt.show()
```

## Conclusion et synthèse des performances

- **DecisionTree** : Très performant sur l'entraînement mais baisse sur validation/test, indiquant un sur-apprentissage. Généralisation limitée.
- **LogisticRegression** : Scores stables et proches entre splits, bonne robustesse et capacité de généralisation.
- **RandomForest** : Excellente performance sur l'entraînement, légère baisse sur validation/test. Moins d'overfitting que DecisionTree, bonne généralisation.
- **SVM** : Performances intermédiaires et cohérentes, dépend du paramétrage. Bon compromis entre complexité et généralisation.
- **XGBoost** : Meilleur modèle, scores élevés et très proches entre splits. Excellente généralisation et robustesse.

**En résumé** : XGBoost et LogisticRegression sont les plus fiables et généralisent le mieux. DecisionTree et RandomForest peuvent sur-apprendre si non régularisés. SVM offre un bon équilibre selon les paramètres. Privilégier les modèles avec des scores proches entre splits pour garantir la robustesse sur de nouvelles données.
